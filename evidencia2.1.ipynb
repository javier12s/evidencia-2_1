```python
class Calculadora:
    def __init__(self, valor_inicial=0):
        self.resultado = valor_inicial

    def sumar(self, num):
        self.resultado += num

    def restar(self, num):
        self.resultado -= num

    def multiplicar(self, num):
        self.resultado *= num

    def dividir(self, num):
        if num != 0:
            self.resultado /= num
        else:
            print("Error: No se puede dividir por cero.")

    def obtener_resultado(self):
        return self.resultado

# Demostración de uso
calculadora = Calculadora(10)

# Realizar algunas operaciones
calculadora.sumar(5)
print("Después de sumar 5, el resultado es:", calculadora.obtener_resultado())

calculadora.multiplicar(2)
print("Después de multiplicar por 2, el resultado es:", calculadora.obtener_resultado())

calculadora.restar(8)
print("Después de restar 8, el resultado es:", calculadora.obtener_resultado())

calculadora.dividir(2)
print("Después de dividir por 2, el resultado es:", calculadora.obtener_resultado())

calculadora.dividir(0)  # Intentar dividir por cero

# Resultado final
print("El resultado final es:", calculadora.obtener_resultado())
```
